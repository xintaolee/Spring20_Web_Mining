```python
# 项目1. A cnblogs 新闻
```

```python
URL_src = { "zh" : "https://www.hurun.net/CN/Article/Details?num=E7190250C866", 
            "en" : "https://www.hurun.net/EN/Article/Details?num=A38B8285034B",}
```

```python
import pandas as pd
df_list = pd.read_html(URL_src['zh'], encoding="utf8", header=0, index_col=0)
# 真幸運有好幾個表
```

```python
from IPython.display import display, HTML
# 快速檢查，看共有幾個表
print (len(df_list))

for df in df_list:
    display(df.head())
```

```python
df_out = dict()
```

## 表格1

```python
i = 0
d = df_list[i].copy().reset_index()
d.head()
```

```python
# 实际上为两个表格串起来：全球独角兽总部_国家, 全球独角兽总部_城市
d.columns = d.iloc[0] # 将第一行内容值作成栏位值
d = d[1:] # 将第一行内容忽然不留
d.head()
```

```python
d.columns
```

```python
import numpy as np
[str(i) if x!=x else x for i,x in enumerate(d.columns)]  #The usual way to test for a NaN is to see if it's equal to itself
```

```python
d.columns = [str(i) if x!=x else x for i,x in enumerate(d.columns)]    #空缺值填0,1,2
d.columns
```

```python
d = d.fillna(0)
d[["0"]] = d[["0"]].astype(int)
d[["4"]] = d[["4"]].astype(int)
```

```python
df_label = "独角兽数_国家"
d.columns.name = df_label
df_out[df_label] = d.iloc[:,[0,1,2]].query("国家!=0")
display(df_out[df_label] )

df_label = "独角兽数_城市"
d.columns.name = df_label
df_out[df_label] = d.iloc[:,[4,5,6]].query("城市!=0")
display(df_out[df_label] )
```

### 檔案文件清單 file manifest
- 输入
url -> pandas.read_html() ->
- 输出
data_sets/hurun_unicorn_{year}.xlsx

```python
fn_out = "data_sets/hurun_unicorn_{yr}.xlsx"

# 分頁
writer = pd.ExcelWriter(fn_out.format(yr=2019))

for p in sorted(list(df_out.keys())) :
    df_out[p].to_excel(writer,"{p}".format(p=p))
    
writer.save()
```
