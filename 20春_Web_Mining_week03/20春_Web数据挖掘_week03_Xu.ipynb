## 数据科学家使用xpath的角度
![](https://qxf2.com/blog/wp-content/uploads/2015/12/Table.png)
* 不挑greedy的策略
    * 求全求不漏
    * 可能有垃圾
    * 使用 // （后代descendant） 而不用  / 子女（children）
    * 使用 **任意**元素/标签  而不用  **指定**元素/标签
* 挑剔ungreedy的策略
    * 求准求数据整齐
    * 可能有漏数据
    * 使用  / 子女（children） 而不用  // （后代descendant）
    * 使用 **指定**元素/标签  而不用  **任意**元素/标签 

### 不挑greedy

````python
# B-B-1 使用 xpath 
r.html.xpath('//a')  # greedy 不挑 所有 <a> 元素标签
````

````python
# B-B-2 使用 xpath 限定 取特定属性
# 注意和 B1 的内容相比, 是不是少了一些? 
# 没有特定属性title就不会被选到
解析后.xpath('//a/@title')  # less greedy 有点挑 <a> 元素标签
````

````python
# B-B-3 使用 xpath  
# greedy 不挑元素/标签  只挑有特定属性title的所有元素
r.html.xpath('//*/@title')  
````

### 挑剔ungreedy

````python
# B-B-4 使用 xpath  # ungreedy 更精準
r.html.xpath('//div[@class="news_title"]/a/@title')
````

````python
# B-B-5 使用 xpath  # ungreedy 最精准
#   xpath 完全没有用 // 也没有用 * 
r.html.xpath('body/div[@class="lin-content"]/div[@class="lin-neiye clearfix"]/div[@class="search_list_right"]/div[@class="ny_content"]/ul[@class="list-ul"]/li/div[@class="news_title"]/a')
````

## 更多xpath
xpath是一门在XML文档（包括html，以樹狀為主的純本文結構文檔）中查找信息的语言

2. 熟悉 [xpath 语法](https://www.w3cschool.cn/xpath/xpath-syntax.html)丶[xpath 节点](https://www.w3cschool.cn/xpath/xpath-nodes.html)
    * 节点
        * 元素丶属性丶文本丶命名空间丶文档（根）结点
    * 节点关系 
        * 父母（parent） vs.先辈（ancestor）
        * 子女（children） vs. 后代（descendant）
        * 同胞（sibling）

````python
def get_e_text_all(_xpath_):
    elements = r.html.xpath(_xpath_)
    # 高級列表推導，看不懂可看另一版本見 # Appendix A
    暂存结果 = list()
    for e in elements:
        outcome = e.xpath("./*/text()")  
        if type(outcome) is str:
            暂存结果.append(outcome)  #有1個字串結果填字串
        elif type(outcome) is list:
            暂存结果.append("".join(outcome))  #有1個以上字串結果填串起來的內容
        else:
            暂存结果.append("")       #无內容也填空
    return(暂存结果)
````

````python
# C-6 
# 难：按參數取prepare requests
# 取出最大页码，并循环获取url

i=0 # 第1页


url = "https://m.liepin.com/zhaopin/pn"+str(i)
print(url)
session = HTMLSession()

data ={
    "compScale":"000",
    "jobkind":"",
    "salaryhigh":"999",
    "compKind":"000",
    "industrys":"000",
    "pubtime":"000",
    "dqs":"050020",  # 省份。例如：广州：dqs=050020
    "keyword":"python", # 行业
    "salarylow":"0",
    "degradeFlag":"0",
}



r = session.get( url , params = data )

# 当前页码
current_pn = int(r.html.xpath('//div/p[contains(@class,"pager-count")]/span[1]/text()')[0])
# 取出最大页码
max_pn = int(r.html.xpath('//div/p[contains(@class,"pager-count")]/span[2]/text()')[0])
print("当前页",current_pn,type(current_pn))
print("最后一页",max_pn,type(max_pn))

# 循环所有页码

content_list = []
数据字典 = {}
dict_xpaths={ 
'职称':    './/ul/li/a[contains(@class,"job-name")]/span[@class="name-text"]/text()', 
'链结':    './/ul/li/a[contains(@class,"job-name")]/@href', 
'薪水':    './/ul/li/a[contains(@class,"job-name")]/following-sibling::span/text()', 
'公司地点':'.//ul/li/time/following-sibling::a/text()',
'公司名称': './/ul/li/a[contains(@class,"company-name")]/text()', 
'时间':    './/ul/li/time/text()', 
#    '公司URL': './/ul/li/a[contains(@class,"company-name")]/@href', 
#    '经验':      './/ul/li/a/following-sibling::text()'
}
for i in range(max_pn):

    url = "https://m.liepin.com/zhaopin/pn"+str(i)
    r = session.get( url , params = data )
    print(url)

    数据字典 ={k:r.html.xpath(v) for k,v in dict_xpaths.items()}

    content_list.append(数据字典)
content_list[0]

````
